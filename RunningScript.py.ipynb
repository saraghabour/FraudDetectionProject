```python
import streamlit as st
from azure.storage.blob import BlobServiceClient
import pickle

# Azure Blob Storage setup
connection_string = "DefaultEndpointsProtocol=https;AccountName=frauddetectiondata;AccountKey=MP76xq6XBfmJnTRYiveWtPg1FlM4WcAyYF6Q3vk93yO+CDJEUlqKFvNXaGIhaK4xUWaqT+Hn/tI7+AStsOXqfg==;EndpointSuffix=core.windows.net"
blob_service_client = BlobServiceClient.from_connection_string(connection_string)

class FraudDetectionApp:
    def __init__(self):
        self.model = self.load_model()

    def load_model(self):
        # Download and load the model from Azure Blob Storage
        self.download_blob("fraudcont", "fraud_detection_model.pkl", "fraud_detection_model.pkl")
        with open("fraud_detection_model.pkl", "rb") as file:
            model = pickle.load(file)
        return model

    def download_blob(self, container_name, blob_name, download_file_name):
        blob_client = blob_service_client.get_blob_client(container="fraudcont", blob='fraud_detection_model.pkl')
        with open(download_file_name, "wb") as download_file:
            download_data = blob_client.download_blob()
            download_file.write(download_data.readall())

    def run(self):
        st.title("Fraud Detection Prediction App")
        st.write("Please enter a number between 0 and 4 to represent the transaction type:")
        st.write("- 0: Cash Out")
        st.write("- 1: Transfer")
        st.write("- 2: Payment")
        st.write("- 3: Debit")
        st.write("- 4: Deposit")

        # User input for features
        transaction_amount = st.number_input("Transaction Amount", min_value=0.0)
        transaction_type = st.number_input("Enter transaction type (0-4):", min_value=0, max_value=4, step=1)
        transaction_time = st.number_input("Transaction Time", min_value=0.0)
        customer_oldbalanceOrg = st.number_input("Customer Old Balance", min_value=0.0)
        customer_newbalanceOrig = st.number_input("Customer New Balance", min_value=0.0)
        dest_oldbalanceOrg = st.number_input("Destination Old Balance", min_value=0.0)
        dest_newbalanceOrig = st.number_input("Destination New Balance", min_value=0.0)
        
        # Predict button
        if st.button("Predict Fraud"):
            features = [
                transaction_amount,
                transaction_type,
                transaction_time,
                customer_oldbalanceOrg,
                customer_newbalanceOrig,
                dest_oldbalanceOrg,
                dest_newbalanceOrig
            ]
            prediction = self.predict_fraud(features)  # Call the prediction method
    
            if prediction == 1:
                st.error("This transaction is likely fraudulent!")
            else:
                st.success("This transaction is not fraudulent.")

    def predict_fraud(self, features):
        # Call your model to make predictions
        return self.model.predict([features])[0]  # Adjust based on your model input

# Run the Streamlit app
if __name__ == "__main__":
    app = FraudDetectionApp()
    app.run()
```
